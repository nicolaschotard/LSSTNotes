# Job history

To have detailed information on all the official productions, or compare productions to each other, check the dedicated [production webpage](http://snf.in2p3.fr/prod/). To get the history of a given production/job in term of job parents/children, have a look at the [data access section](DataAccess.html#process-job-history) on process/job history. You will find below the main new features of a production/IDR compared to the previous one.

## Productions 

### SNF-0203-NEWYORK

* Info
    * "Flux calibration" production [02/2016]
    * Web pages: [zoo](htpp://snf.in2p3.fr/zoo/SNF-0203-NEWYORKa), [prod](htpp://snf.in2p3.fr/prod/SNF-0203-NEWYORKa), [slides](https://snf-doc.lbl.gov/twiki/pub/Conf/VideoConf20160218/prod_status_02-18-2016.pdf).
    * History: http://snf.in2p3.fr/history/SNF-0203-NEWYORKf/
* Based on
    * PAT: SNF-0203-NEWYORKa
    * PFC: SNF-0203-NEWYORKf
    * PFS: SNF-0203-NEWYORKx
    * PMS: SNF-0203-NEWYORKm
    * PES: SNF-0203-ES0203e, 
    * MFR: SNF-0203-NEWMFR
    * PCG: SNF-0116-Cu0116c, SNF-0202-Cu0202c, SNF-0203-Cu0203c, SNF-0200-Cu0200c, SNF-0201-Cu0201c 
    * PFQ: SNF-0201-Pr0201z, SNF-0202-Pr0202z, SNF-0203-Pr0203z
    * PSA (synthetic arcs): SNF-0201-PSA
    * See the history webpage above for details
    
* New wrt. SNF-0203-NEWMEX
    * Using a full new MFR production: SNF-0203-NEWMFR
        * +571 MFRs compared to SNF-0202-NEWMFR (still 2100 missing)
        * Much less attrition at the target level and at the rescaled MFR level
        * Cut was 3 expsosures for MFR computation, it is now 2.
        * Made sure that no MFR was computed using a “shifted” final reference
    * Added nights of January 2016 (2 nights, 1 photo)
    * XNIGHT flux calibrated problem solved (see [mantis \#1548](https://projects.lbl.gov/mantis/view.php?id=1548))
        * Arc prod correctly taken into account
        * Fit problem at the flux calibration level solved
        * All arcs and their children flagged
            * Fclass/XFclass = 4/901 for artificial arcs
            * Flag: “ARTIFICIAL_ARC”
    * Relative flux calibration active (“PFC_RELFLX”)
* Associated host-subtracted productions
    * **SNF-0203-ALLEG2a**
        * Info
            * "Host subtraction" production [02/2016]
            * Web pages: [zoo](htpp://snf.in2p3.fr/zoo/SNF-0203-ALLEG2a), [prod](htpp://snf.in2p3.fr/prod/SNF-0203-ALLEG2a), [slides](https://snf-doc.lbl.gov/twiki/pub/Conf/VideoConf20160225/prod_status_02-25-2016.pdf).
        * Based on
            * SNF-0203-ALLEGe (PES)
            * SNF-0203-ALLEG_R/Bb (PCF)
            * SNF-0203-NEWYORKf (PFC)
        * New wrt. SNF-0203-ALLEG
            * plan_analyze_timeseries production using SALT2 version 2-4-0.
            * 2-2-2b was used in SNF-0203-ALLEGa and all previous production.
    * **SNF-0203-ALLEGa**
        * Info
            * "Host subtraction" production [02/2016]
            * Weg pages: [zoo](htpp://snf.in2p3.fr/zoo/SNF-0203-ALLEGa), [prod](htpp://snf.in2p3.fr/prod/SNF-0203-ALLEGa).
        * Based on
            * SNF-0203-ALLEGe (PES)
            * SNF-0203-ALLEG_R/Bb (PCF)
            * SNF-0203-NEWYORKf (PFC)
        * New wrt. SNF-0203-CABALLO
            * SN-subtracted cube now produced (Fclass/XFclass)
            * SALT2 fits now includes "PCF_XNIGHTS" and "ARTICIAL_ARC" flagged spectra, which actually makes some fits better, while some other get worse:
                * Example for [worst fit](http://snf.in2p3.fr/lc/SNF-0203-ALLEGa,SNF-0203-CABALLOa/PTF12hwb/)
                * Example for [better fit](http://snf.in2p3.fr/lc/SNF-0203-ALLEGa,SNF-0203-CABALLOa/SN2005bg/)
